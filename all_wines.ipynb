```python
csv_path = "Resources/winemag-1-800.csv"
wines_1_800 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-801-1600.csv"
wines_801_1600 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-1601-2400.csv"
wines_1601_2400 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-2401-3200.csv"
wines_2401_3200 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-3201-4000.csv"
wines_3201_4000 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-4001-4800.csv"
wines_4001_4800 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-4801-5600.csv"
wines_4801_5600 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-5601-6400.csv"
wines_5601_6400 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-6401-7200.csv"
wines_6401_7200 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-7201-8000.csv"
wines_7201_8000 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-8001-8800.csv"
wines_8001_8800 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-9601-10400.csv"
wines_9601_10400 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-10401-11200.csv"
wines_10401_11200 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-11201-12000.csv"
wines_11201_12000 = pd.read_csv(csv_path,encoding = 'utf-8')

csv_path = "Resources/winemag-12001-12800.csv"
wines_12001_12800 = pd.read_csv(csv_path,encoding = 'utf-8')
```

```python
all_wines = pd.concat([wines_1_800, wines_801_1600, wines_1601_2400,wines_2401_3200,wines_3201_4000,wines_4001_4800,wines_4801_5600,wines_5601_6400,wines_6401_7200,wines_7201_8000,wines_8001_8800,wines_9601_10400,wines_10401_11200,wines_11201_12000,wines_12001_12800])
all_wines_df = all_wines.reset_index()        
all_wines_df.index.name = 'ID'
all_wines_df.drop('index', axis=1, inplace=True)
all_wines_df
```

```python
all_wines_df.to_csv("Output/all_wines.csv",encoding='utf-8-sig', index=False, header=True)
```

```python
engine = create_engine('sqlite:///SQL/allwines.sql', encoding="utf8", echo=False)
all_wines_df.to_sql('all_wines', con=engine)
total_wines = engine.execute("SELECT * FROM all_wines").fetchall()
```
